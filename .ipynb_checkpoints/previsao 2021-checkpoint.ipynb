```python
#grafico

estacao=['45590000','45840000','45960001'
,'46543000','46590000']

q_noventa=[23.1506,54.35319,138.6621,
           32.4929,38.1553]

vsup2015=[0.001898148,15.36604167,40.44226512,
          21.93714466,27.58587037]



vsup2020=[1.028923611,17.68288264,48.99635454,
          10.75447917,13.99731377]

nome=['Correntina - Correntina', 'Formoso - Gatos', 'Corrente - Porto Novo',
     'de Ondas - Fazendo Redenção', 'Branco - São Sebastião']


for i in range(len(estacao)):
    telemetria= pd.read_csv(estacao[2]+'_2021.csv',sep=';')
    telemetria=pd.DataFrame(telemetria)
    telemetria['data'] = pd.to_datetime(telemetria['data'],format='%d/%m/%Y')
    telemetria['dia'] = telemetria['data'].dt.dayofyear

    vazao_zero=np.array(telemetria['vazao diaria'])
    y_feito= telemetria['vazao abril']
    y_feito_150=telemetria['vazao maio']
    tempo=np.array(telemetria['dia'])
    min_abril=telemetria['vazao abril'].min()
    min_maio=telemetria['vazao maio'].min()
    
    fig, ax = pltp.subplots()
    pltp.rc("font",family="Times New Roman")
    pltp.title('Estação '+estacao[2]+' - Rio '+nome[2], fontsize=12, fontname="Times New Roman")
   
    pltp.scatter(tempo, vazao_zero, label='Vazão Observada', s=10,color= 'b')
    pltp.plot(tempo, y_feito, label='Previsão de vazão iniciada em abril', color='red')
    pltp.plot(tempo, y_feito_150, label='Previsão de vazão iniciada em maio', color='yellow')
   
    pltp.axhline(y=q_noventa[2], label='Q90 (m³/s)', linestyle='-',linewidth=1.3,color='black')
    pltp.axhline(y=vsup2015[2], label='Vazão superficial outorgada em 2015 (m³/s)',linestyle='--',linewidth=1.3, color='black')
    pltp.axhline(y=vsup2020[2], label='Vazão superficial outorgada em 2020  (m³/s)',linestyle='--',linewidth=1.3, color='maroon')
     
   
    pltp.ylabel('Vazão (m³/s)')
    handles, labels = ax.get_legend_handles_labels()                                            #legendas aparecerem no plot
    lgd = pltp.legend(handles, labels, loc='upper center',bbox_to_anchor=(0.50, - 0.2))
    ax.annotate('vazão min prevista abril= %6.3f' %min_abril, xy = (180,120)) 
    ax.annotate('vazão min prevista maio = %6.3f' %min_maio, xy = (180,110))

    ax.set_yticks([0,30,60,90,120,150
                  ])
    lista_mes= ['Mar','Apr','May', 'Jun', 'Jul','Aug','Sep', 'Oct','Nov','Dec']
    ax.set_xticks([61,92,122,153,183,214,245,275,306,336])
    ax.set_xticklabels(lista_mes,rotation=0)
    pltp.xlabel('Meses')
    pltp.savefig(estacao[2]+'_2021_previsao.png',dpi=500, bbox_inches='tight',pad_inches=0.2)
```
